```python
#Importing Libraries
import warnings
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.svm import SVR 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
```

```python
warnings.filterwarnings("ignore")
```

```python
#Loading the Dataset
dataset = pd.read_csv("prep.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
```

```python
dataset
```

```python
independent = dataset.drop("classification_yes",axis = 1)
dependent = dataset["classification_yes"]
```

```python
independent
```

```python
dependent
```

```python
#Finding best Features Using RFE
n = 4
rfe_list = []
lr = LinearRegression()
svr_linear = SVR(kernel = 'linear')
#svr_non_liner = SVR(kernel = 'rbf') 
dt = DecisionTreeRegressor(random_state = 0)
rf = RandomForestRegressor(n_estimators = 10, random_state = 0)
 
#regression_models = [lr, svr_linear, svr_non_liner, dt,rf]  

regression_models = [lr, svr_linear, dt,rf]  

for model in regression_models:
    print(f"MODEL : {model}") 
    rfe_obj = RFE(model,n_features_to_select = n)
    rfe_model = rfe_obj.fit(independent,dependent)
    rfe_feature = rfe_model.transform(independent)
    rfe_list.append(rfe_feature)
    print(f"Completed")
    
rfe_list  
```

```python
lr_acc = []
svr_linear_acc = []
svr_non_linear_acc = []
dt_acc = []
rf_acc = []
 
```

```python
for i in rfe_list:
    x_train, x_test, y_train, y_test = train_test_split(i,dependent,test_size = 0.20, random_state = 0)

    #Linear Regresssion
    lr_regression = LinearRegression()
    lr_regression.fit(x_train, y_train)
    y_pred = lr_regression.predict(x_test)  
    lr_acc.append(r2_score(y_test,y_pred))
   
    #SVM Linear
    svc_regressor =  SVR(kernel = 'linear')
    svc_regressor.fit(x_train, y_train)
    y_pred = svc_regressor.predict(x_test)  
    svr_linear_acc.append(r2_score(y_test,y_pred))  
     
    #SVM Non Linear
    svc_n_regressor =  SVR(kernel = 'rbf')
    svc_n_regressor.fit(x_train, y_train)
    y_pred = svc_n_regressor.predict(x_test)  
    svr_non_linear_acc.append(r2_score(y_test,y_pred))  
     
    #Random Forest
    rf_classifier =  RandomForestRegressor(n_estimators = 10, random_state = 0)
    rf_classifier.fit(x_train, y_train)
    y_pred = rf_classifier.predict(x_test) 
    rf_acc.append(r2_score(y_test,y_pred))  

    #Decistion Tree
    dt_regressor =DecisionTreeRegressor(random_state = 0)
    dt_regressor.fit(x_train, y_train)
    y_pred = dt_regressor.predict(x_test)  
    dt_acc.append(r2_score(y_test,y_pred))  
```

```python
rfe_data = pd.DataFrame(index = ["Linear","SVC","Random","DecisionTree"],columns = ["Linear","SVM_Linear","Random","DecisionTree"])
```

```python
rfe_data
```

```python
for index,col in enumerate(rfe_data.index):    
    rfe_data["Linear"][col]          = lr_acc[0] 
    rfe_data["SVM_Linear"][col]             = svr_linear_acc[0]
    rfe_data["Random"][col]          = rf_acc[0]
    rfe_data["DecisionTree"][col]  = dt_acc[0]

 
```
