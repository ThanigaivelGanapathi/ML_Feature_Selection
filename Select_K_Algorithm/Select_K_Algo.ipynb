```python
#Importing Libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
#Loading the Dataset
dataset = pd.read_csv("Prep.csv")
```

```python
dataset
```

```python
#Converting Categorical data to Numerical Data
dataset = pd.get_dummies(dataset,dtype = int,drop_first = True)
```

```python
dataset
```

```python
#Splitting Dependent and Independent Variables
independent = dataset.drop("classification_yes",axis = 1)
dependent = dataset["classification_yes"]
```

```python
independent
```

```python
dependent
```

```python
#Feature Selection Using Select K Best Algorithm
kBest = SelectKBest(score_func=chi2, k=5)
kBestAlgo = kBest.fit(independent,dependent)
features = kBest.transform(independent)
features
```
