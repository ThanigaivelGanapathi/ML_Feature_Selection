```python
#Importing Libraries
import pandas as pd
import numpy as np
import warnings
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
   
```

```python
warnings.filterwarnings('ignore')
```

```python
#Loading the Dataset
dataset = pd.read_csv("Prep.csv")
```

```python
dataset
```

```python
#Converting Categorical data to Numerical Data
dataset = pd.get_dummies(dataset,dtype = int,drop_first = True)
```

```python
dataset
```

```python
#Splitting Dependent and Independent Variables
independent = dataset.drop("classification_yes",axis = 1)
dependent = dataset["classification_yes"]
```

```python
independent
```

```python
dependent
```

```python
#Feature Selection Using Select K Best Algorithm
kBest = SelectKBest(score_func=chi2, k=7)
kBestAlgo = kBest.fit(independent,dependent)
features = kBest.transform(independent)
features
```

```python
#Splitting Training and Testing Dataset
x_train, x_test, y_train, y_test = train_test_split(features,dependent,test_size = 0.20,random_state = 0)
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test) 
```

```python
linear=[]
svml=[]
svmnl=[]
des=[]
rf=[]
```

```python
for i in features:        
    #Linear Regression
    linear_regression = LinearRegression()
    linear_regression.fit(x_train,y_train)
    y_pred = linear_regression.predict(x_test)
    linear.append(r2_score(y_test,y_pred)) 

    #SVM Linear
    svm_linear = SVR(kernel = 'linear')
    svm_linear.fit(x_train,y_train)
    y_pred = svm_linear.predict(x_test)
    svml.append(r2_score(y_test,y_pred)) 

    #SVM Non Linear
    svm_non_linear = SVR(kernel = 'rbf')
    svm_non_linear.fit(x_train,y_train)
    y_pred = svm_non_linear.predict(x_test)
    svmnl.append(r2_score(y_test,y_pred)) 

    #Decision Tree
    decision_tree = DecisionTreeRegressor(random_state = 0)
    decision_tree.fit(x_train,y_train)
    y_pred = decision_tree.predict(x_test)
    des.append(r2_score(y_test,y_pred)) 

    #Random Forest
    random_forest = RandomForestRegressor(n_estimators = 10, random_state = 0)
    random_forest.fit(x_train,y_train)
    y_pred = random_forest.predict(x_test)
    rf.append(r2_score(y_test,y_pred)) 

```

```python
dataframe=pd.DataFrame(index=['ChiSquare'],columns=['Linear','SVMl','SVMnl','Decision','Random']) 
for number,index in enumerate(dataframe.index):        
    dataframe['Linear'][index]=linear[number]       
    dataframe['SVMl'][index]=svml[number]
    dataframe['SVMnl'][index]=svmnl[number]
    dataframe['Decision'][index]=des[number]
    dataframe['Random'][index]=rf[number]
```
